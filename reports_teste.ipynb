```python
import matplotlib.pyplot as plt
import numpy as np

classes = [f"C{i}" for i in range(1, 17)]

precision = [
    0.99, 0.98, 0.87, 0.98, 1.00, 0.99, 0.97, 1.00,
    0.98, 0.73, 0.82, 1.00, 0.98, 1.00, 1.00, 1.00
]

recall = [
    1.00, 0.62, 0.79, 1.00, 1.00, 0.98, 1.00, 0.98,
    1.00, 0.98, 0.86, 0.98, 1.00, 1.00, 1.00, 0.97
]

f1_score = [
    0.99, 0.76, 0.83, 0.99, 1.00, 0.99, 0.98, 0.99,
    0.99, 0.84, 0.84, 0.99, 0.99, 1.00, 1.00, 0.99
]



x = np.arange(len(classes))
width = 0.25

plt.figure(figsize=(14, 6))

plt.bar(x - width, precision, width, label="Preciso")
plt.bar(x, recall, width, label="Sensibilidade")
plt.bar(x + width, f1_score, width, label="F1-score")

plt.xticks(x, classes)
plt.title("Mtricas por Classe - Classificador Federdo (=0.5)", 
          fontsize=20, fontweight='bold')
plt.xlabel("Classes", fontsize=20, fontweight='bold')
plt.ylabel("Score", fontsize=20, fontweight='bold')
plt.ylim(0.55, 1.01)
plt.legend(loc='lower right', fontsize=14)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tick_params(axis='both', labelsize=16)

plt.tight_layout()
plt.savefig("resultados_federado/federado_report_alpha_0.5.png", dpi=150, bbox_inches='tight')
plt.show()
```
